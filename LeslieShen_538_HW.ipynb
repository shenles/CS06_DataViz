```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style

marriage_data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/marriage/both_sexes.csv')
marriage_data.head()
```

```
# figures represent share of population that has never been married
# need to subtract each number from 1 to get the share of the population that HAS been married
marriage_data['hs2534'] = marriage_data.apply(lambda row: 1 - row.HS_2534, axis=1)
marriage_data['sc2534'] = marriage_data.apply(lambda row: 1 - row.SC_2534, axis=1)
marriage_data['ba2534'] = marriage_data.apply(lambda row: 1 - row.BAp_2534, axis=1)
```

```
education_25to34 = pd.Index(['hs2534', 'sc2534', 'ba2534'])
```

```
style.use('fivethirtyeight')
colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255], [213/255,94/255,0], [0,114/255,178/255]]

marriage_plot = marriage_data.plot(x='year', y=education_25to34, figsize=(8, 10), color=colors, legend=False)
marriage_plot.axhline(y=0, alpha=0.7, linewidth=1.5, color='black')
marriage_plot.set_ylim(top=1.1)
marriage_plot.set_yticklabels(labels = [-10, '0  ', '20  ', '40  ', '60  ', '80  ', '100%'])
marriage_plot.xaxis.label.set_visible(False)
marriage_plot.text(x=1952.5, y=1.2, s='Marriage Rates By Education', fontsize=26, weight='bold', alpha =.75)
marriage_plot.text(x=1952.5, y=1.15, s='Ages 25 to 34', fontsize=19, alpha=.85)
marriage_plot.text(x=1985, y=.56, s='College graduates', color=colors[2], weight='bold', rotation=-6, backgroundcolor='#f0f0f0')
marriage_plot.text(x=1974, y=.78, s='High school or less', color=colors[0], weight='bold', rotation=-31, backgroundcolor='#f0f0f0')
marriage_plot.text(x=1995, y=.68, s='Some college', color=colors[1], weight='bold', rotation=-12, backgroundcolor='#f0f0f0')
marriage_plot.text(x=1952.5,
              y=-0.14,
              s="""CS06 Data Viz with Python """
                + """                        """
                + """Based on Census Bureau data""",
              fontsize=15,
              color='#f0f0f0',
              backgroundcolor='grey'
              )
plt.show()
```
